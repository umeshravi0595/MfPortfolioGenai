```python
%sql
CREATE CATALOG IF NOT EXISTS retail_analytics;
CREATE SCHEMA IF NOT EXISTS retail_analytics.pizza;
CREATE VOLUME IF NOT EXISTS retail_analytics.pizza.raw_data;
```

```python
%sql
-- CREATE VOLUME IF NOT EXISTS retail_analytics.restaurants.bronze;
```

```python
from pyspark.sql.functions import *

sales_bronze = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/retail_analytics/pizza/raw_data/Pizza_Sale.csv")
    .withColumn("ingest_ts", current_timestamp())
)

sales_bronze.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("retail_analytics.pizza.bronze_sales")
```

```python
ingredients_bronze = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/retail_analytics/pizza/raw_data/Pizza_ingredients.csv")
    .withColumn("ingest_ts", current_timestamp())
)

ingredients_bronze.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("retail_analytics.pizza.bronze_ingredients")
```
